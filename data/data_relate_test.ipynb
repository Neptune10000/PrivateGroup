# 特征分析
## 1.不同日期点击量的变化


```python
train['clickDate'] = train['clickTime'].apply(lambda x : x/10000)
```

```python
train.groupby(['clickDate']).sum()['label'].plot(figsize=(12,6))
```

18 19 20三天是异常数据，使用21-29数据比较合适

## 2.不同creativeID的变化

```python
train.groupby(['creativeID']).sum()['label'].plot(figsize=(12,6))
```

creativeID对于是否回流影响很大，应该作为特征使用

## 3.不同用户的变化

```python
train.groupby(['userID']).sum()['label'].plot(figsize=(12,6))
```

可以看出数据主要集中在0-2，userID对于判断是否回流作用不大，可以不用这个特征。
